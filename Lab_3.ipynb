```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# For better plots
sns.set(style='whitegrid')

# Load the wine dataset
wine = datasets.load_wine()

# Convert to a pandas DataFrame for easier exploration
data = pd.DataFrame(wine.data, columns=wine.feature_names)
data['target'] = wine.target

# Display the first few rows
data.head()
```

```python
# Check the shape of the dataset
print(f"Dataset shape: {data.shape}")

# Check for missing values
print("\nMissing values:\n", data.isnull().sum())

# Summary statistics
print("\nSummary statistics:\n")
display(data.describe())

# Class distribution
print("\nClass distribution:")
display(data['target'].value_counts())
```

```python
# Separate features and target
X = data.drop('target', axis=1)
y = data['target']

# Apply Z-score normalization (Standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for easier handling
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Display the first few rows of the standardized data
X_scaled_df.head()
```

```python
# Import KMeans and evaluation metrics
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, adjusted_rand_score

# Initialize KMeans with k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# Fit the model
kmeans.fit(X_scaled_df)

# Get the cluster labels
kmeans_labels = kmeans.labels_

# Display first 10 cluster labels
print("Cluster Labels:", kmeans_labels[:10])
```

```python
# Calculate Silhouette Score
silhouette = silhouette_score(X_scaled_df, kmeans_labels)
print(f"Silhouette Score: {silhouette:.4f}")
```

```python
# Calculate Adjusted Rand Index
ari = adjusted_rand_score(y, kmeans_labels)
print(f"Adjusted Rand Index (ARI): {ari:.4f}")
```

```python
# Import KMedoids from scikit-learn-extra
from sklearn_extra.cluster import KMedoids

# Initialize K-Medoids with k=3
kmedoids = KMedoids(n_clusters=3, random_state=42)

# Fit the model
kmedoids.fit(X_scaled_df)

# Get cluster labels
kmedoids_labels = kmedoids.labels_

# Display first 10 cluster labels
print("K-Medoids Cluster Labels:", kmedoids_labels[:10])
```

```python
# Calculate Silhouette Score for K-Medoids
silhouette_kmedoids = silhouette_score(X_scaled_df, kmedoids_labels)
print(f"Silhouette Score (K-Medoids): {silhouette_kmedoids:.4f}")
```

```python
# Calculate Adjusted Rand Index for K-Medoids
ari_kmedoids = adjusted_rand_score(y, kmedoids_labels)
print(f"Adjusted Rand Index (ARI) (K-Medoids): {ari_kmedoids:.4f}")
```

```python
from sklearn.decomposition import PCA

# Apply PCA to reduce data to 2 components for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_df)

# Put PCA results into a DataFrame for convenience
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['KMeans Cluster'] = kmeans_labels
pca_df['KMedoids Cluster'] = kmedoids_labels

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot K-Means
axes[0].scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['KMeans Cluster'], cmap='viridis', s=50)
axes[0].scatter(
    pca.transform(kmeans.cluster_centers_)[:, 0],
    pca.transform(kmeans.cluster_centers_)[:, 1],
    c='red', marker='X', s=200, label='Centroids'
)
axes[0].set_title('K-Means Clustering')
axes[0].legend()

# Plot K-Medoids
axes[1].scatter(pca_df['PC1'], pca_df['PC2'], c=pca_df['KMedoids Cluster'], cmap='viridis', s=50)
axes[1].scatter(
    pca.transform(kmedoids.cluster_centers_)[:, 0],
    pca.transform(kmedoids.cluster_centers_)[:, 1],
    c='red', marker='X', s=200, label='Medoids'
)
axes[1].set_title('K-Medoids Clustering')
axes[1].legend()

plt.show()
```
